```python
#import needed packages
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
# Load and display the first few rows of the dataset to understand its structure
vo = pd.read_csv('vo - Sheet1.csv')

# Display the first few records to understand the dataset
o3 = pd.read_csv('o3 - Sheet1.csv')

no = pd.read_csv('no - Sheet1.csv')
```

```python
no['type']='no'
```

```python
vo['type']='vo'
```

```python
o3['type']='o3'
```

```python
vo
```

```python
combine_df = pd.concat([vo, o3, no], axis=0, ignore_index=True)
```

```python
combine_df
```

```python

melted_df = pd.melt(combine_df, id_vars=['code', 'iso', 'country', 'type'], var_name='year', value_name='air_data')

# Convert 'year' column to datetime type
melted_df['year'] = pd.to_datetime(melted_df['year'], format='%Y')

# Create a pivot table
pivot_table = melted_df.pivot_table(index=['code', 'iso', 'country', 'type', 'year'], values='air_data')

# Reset the index to make 'year' a regular column
pivot_table = pivot_table.reset_index()

# Display the pivot table
pivot_table
```

```python
pivot_table.to_csv('combine.csv')
files.download("combine.csv")
```
